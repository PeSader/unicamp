# Ressonâncias em Tubos

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np 
import math


x = np.arange(1,10000) # frequência
v = 340.27 # velocidade do som (m/s)
```

## Tubo Pequeno (L = 22.5cm)

```python
D = 0.037 # diâmetro do tubo pequeno (m)
L = 0.225 # comprimento do tubo pequeno (m)

# Lefetivo = 0.6*r*2 = 0.6*D
Lefetivo = L + 0.6*D
L = Lefetivo

print('Lefetivo =', Lefetivo)
```

```python
RATIOMAX = 5
RATIOMIN = 1

FMIN = v/(L*RATIOMAX)
FMAX = v/(L*RATIOMIN)
print(FMIN, FMAX)

y = v / (x*L)
y1 = RATIOMIN
y2 = RATIOMAX
```

```python
# Pontos para destacar em tubo aberto-fechado pequeno
x_fechado = []
y_fechado = []
for n in range(1, 9):
    xf = (v*(2*n-1))/4*L
    x_fechado.append(xf)

    yf = v / (x_fechado[n-1]*L)
    y_fechado.append(yf)

print('len(x_fechado):', len(x_fechado))
print(x_fechado)
print()
print('len(y_fechado):', len(y_fechado))
print(y_fechado)
    
```

```python
fig1, ax1 = plt.subplots()
fig1.set_figwidth(16)
fig1.set_figheight(8)

xmax = 2500
# xmax = 10000

ax1.plot(x, y)
ax1.set_ylim(0, 15)
ax1.set_xlim(xmax=xmax)

area_labels=['Área I','Área II','Área III']

#text_height = 14
#plt.axvline(FMIN, alpha=0.3, color='black')
#plt.axvline(FMAX, alpha=0.3, color='black')

ax1.plot(x_fechado, y_fechado, marker='s', linestyle = '', color='C0', label='Frequências harmônicas')

threshold_max = RATIOMAX
ax1.fill_between(x, 0, 1, where=y > threshold_max,
                color='C1', alpha=0.15, transform=ax1.get_xaxis_transform(), label=area_labels[0])
#ax1.text((FMIN/2), text_height, area_labels[0])

ax1.axvspan(FMIN, FMAX, alpha=0.10, color='C2', label=area_labels[1])
#ax1.text((FMIN+FMAX)/2, text_height, area_labels[1])

threshold_min = RATIOMIN
ax1.fill_between(x, 0, 1, where=y < threshold_min,
                color='C4', alpha=0.10, transform=ax1.get_xaxis_transform(), label=area_labels[2])
#ax1.text((FMAX+xmax)/2.5, text_height, area_labels[1])


xticks = np.append(ax1.get_xticks(), [FMAX, FMIN, 100])
ax1.set_xticks(xticks)
yticks = np.append(ax1.get_yticks(), [1])
ax1.set_yticks(yticks)
degrees = 40
plt.xticks(rotation=degrees)

plt.legend(loc='best')

handles, labels = plt.gca().get_legend_handles_labels()
order = [0,2,1,3]
plt.legend([handles[idx] for idx in order],[labels[idx] for idx in order])


ax1.set_xbound(lower=100, upper=xmax)
ax1.set_ylabel('λ / L',fontsize=12)
ax1.set_xlabel('Frequência (Hz)',fontsize=12)
ax1.set_title('Razão Entre Comprimento de Onda (λ) e Comprimento do Tubo Pequeno aberto-fechado (L) em Função da Frequência Sonora\n', fontsize=15)

# plt.savefig('tubo_pequeno_aberto_fechado.png')
```

```python
# Pontos para destacar em tubo aberto-aberto pequeno
x_aberto = []
y_aberto = []
for n in range(1, 16):
    xa = (v*n/2*L)
    x_aberto.append(xa)

    ya = v / (x_aberto[n-1]*L)
    y_aberto.append(ya)

print('len(x_aberto):', len(x_aberto))
print(x_aberto)
print()
print('len(y_aberto):', len(y_aberto))
print(y_aberto)
    
```
